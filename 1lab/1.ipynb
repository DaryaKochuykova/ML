```python
adj_price = df["Close"]
print("Avg:", adj_price.mean())
print("Median:", adj_price.median())
print("Max:", adj_price.max())
print("Min:", adj_price.min())
print("Delta:", adj_price.max()-adj_price.min())
print("Div:", adj_price.std())
print("Dispersion:", adj_price.std()**2)
plt.show()
```
